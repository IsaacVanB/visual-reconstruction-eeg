```python
# TRAIN CLASSIFIER
%%bash


for iter in 1 2 3; do
  python scripts/train_eeg_classifier.py \
    --dataset-root /content/eeg_data/datasets \
    --output-dir /content/drive/MyDrive/eeg_classifier_runs \
    --model-architecture eegnet \
    --epochs 40 \
    --lr 7e-4 \
    --l1-weight 1e-4 \
    --eegnet-dropout 0.1 \
    --eegnet-f1 16 \
    --eegnet-d 4 \
    --eeg-window-pre-ms 0 \
    --eeg-window-post-ms 450 \
    --run-change-note "\
      run $iter\
      "
done
```
